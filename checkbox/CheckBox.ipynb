```python
'''
June 10. Threshold is 150 with percentage as 10%. Accuracy of 85%
'''
import numpy as np
import cv2
import math
from scipy import ndimage

def adjust_gamma(image, gamma):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
 
    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)

def apply_brightness_contrast(input_img, brightness, contrast):
    if brightness != 0:
        if brightness > 0:
            shadow = brightness
            highlight = 255
        else:
            shadow = 0
            highlight = 255 + brightness
        alpha_b = (highlight - shadow)/255
        gamma_b = shadow

        buf = cv2.addWeighted(input_img, alpha_b, input_img, 0, gamma_b)
    else:
        buf = input_img.copy()

    if contrast != 0:
        f = 131*(contrast + 127)/(127*(131-contrast))
        alpha_c = f
        gamma_c = 127*(1-f)

        buf = cv2.addWeighted(buf, alpha_c, buf, 0, gamma_c)
        
    gray = cv2.cvtColor(buf, cv2.COLOR_BGR2GRAY)

    return adjust_gamma(gray,1.25)

def checkbox(x):
    checked = False

    path = str(x) + ".jpg"
    print (path)
    
    img_temp = cv2.imread(path)
    
    img = apply_brightness_contrast(img_temp, -40, 110)
    cv2.imwrite("testing.jpg", img)
    
    height, width = img.shape[:2]
    total = height * width * 3
    
    n_black_pix = np.sum(img < 30)
#     print (n_black_pix)
    
    percentage = n_black_pix/total
    print (percentage)

    if (percentage > 0.013):
        checked = True
        
    return (checked)

count = 0
total = 0
for x in range (1,53):
    positive = checkbox(x)
    total += 1
    print (positive)
    if (positive):
        count += 1
        
    if (x == 14 or x == 28 or x == 29 or x == 30 or x == 33 or x == 39 or x==40 or x == 50 or x == 51):
        if (positive == True):
            print ("False Positive")
        else:
            count += 1
    print ()
    
print (count)
print (total)
#     print ()
    
# cv2.imshow()
# cv2.imshow('test.jpg',img)
# cv2.waitKey(0)
```
